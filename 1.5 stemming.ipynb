## 1.5 Stemming

### What is Stemming?
Stemming reduces words to their root form by removing suffixes. It's a crude method that chops off endings to find the stem.

### Why Use Stemming?
- Reduces vocabulary size
- Groups similar words together
- Improves text matching and search

### Common Stemming Algorithms
- **Porter Stemmer**: Most popular, rule-based
- **Snowball Stemmer**: Improved version of Porter
- **Lancaster Stemmer**: More aggressive, can over-stem

### Limitations
- Can produce non-words (e.g., "running" → "run")
- Not always linguistically correct
- Better alternative: Lemmatization

```python
from nltk.stem import PorterStemmer
```

```python
# create stemmer
ps = PorterStemmer()
```

```python
connect_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connects']
for t in connect_tokens:
    print(t, " : ", ps.stem(t))
```

```python
learn_tokens = ['learned', 'learning', 'learn', 'learns', 'learner', 'learners']
for t in learn_tokens:
    print(t, " : ", ps.stem(t))
```

```python
likes_tokens = ['likes', 'better', 'worse']
for t in likes_tokens:
    print(t, " : ", ps.stem(t))
```
