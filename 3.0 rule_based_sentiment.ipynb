# 3.0 Rule-Based Sentiment Analysis

## TextBlob
Simple, beginner-friendly library for sentiment polarity (0 to 1 scale) and subjectivity. Uses pre-built lexicon of sentiment words. Fast and lightweight, no training required.

## VADER
Specialized for social media & informal text. Handles emoticons, emojis, capitalization, punctuation. Returns compound score (-1 to 1). Better for nuanced sentiment detection.

## Comparison
- **TextBlob**: General-purpose, simple usage, good baseline
- **VADER**: Social media optimized, captures intensity & context, handles informal language

## Strengths & Limitations
✓ No training data needed
✓ Interpretable results
✗ Can't learn domain-specific sentiment
✗ Struggles with sarcasm, complex context
✗ Limited to predefined sentiment words

## Use Cases
- Customer reviews quick analysis
- Social media monitoring
- Baseline for ML models
- Real-time sentiment detection

### Textblob

```python
from textblob import TextBlob
```

```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

```python
print(sentence_1)
sentiment_score = TextBlob(sentence_1)
print(sentiment_score.sentiment.polarity)
```

```python
print(sentence_2)
sentiment_score_2 = TextBlob(sentence_2)
print(sentiment_score_2.sentiment.polarity)
```

```python
print(sentence_3)
sentiment_score_3 = TextBlob(sentence_3)
print(sentiment_score_3.sentiment.polarity)
```

```python
print(sentence_4)
sentiment_score_4 = TextBlob(sentence_4)
print(sentiment_score_4.sentiment.polarity)
```

### VADER

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
vader_sentiment = SentimentIntensityAnalyzer()
```

```python
print(sentence_1)
print(vader_sentiment.polarity_scores(sentence_1))
```

```python
print(sentence_2)
print(vader_sentiment.polarity_scores(sentence_2)) 
```

```python
print(sentence_3)
print(vader_sentiment.polarity_scores(sentence_3)) 
```

```python
print(sentence_4)
print(vader_sentiment.polarity_scores(sentence_4)) 
```
