## 1.4 Tokenizing Text

### What is Tokenization?
Tokenization is the process of breaking down text into smaller units called tokens. These tokens can be words, phrases, symbols, or other meaningful elements that serve as the basic building blocks for NLP tasks.

### Why Tokenize Text?
- **Foundation for NLP**: Most NLP algorithms require text to be broken into tokens
- **Feature Extraction**: Enables creation of numerical representations
- **Text Analysis**: Allows analysis at word/phrase level
- **Preprocessing**: Essential step before stemming, lemmatization, or vectorization

### Types of Tokenization

#### 1. Word Tokenization
- Splits text into individual words
- Handles punctuation and whitespace
- Example: "Hello, world!" → ["Hello", ",", "world", "!"]

#### 2. Sentence Tokenization
- Splits text into sentences
- Uses punctuation marks as delimiters
- Example: "Hello world. How are you?" → ["Hello world.", "How are you?"]

#### 3. Subword Tokenization
- Breaks words into smaller units (subwords)
- Useful for handling out-of-vocabulary words
- Common in modern language models (BPE, WordPiece)

### Challenges in Tokenization
- **Contractions**: "don't" → "do" + "n't" or "don't"
- **Abbreviations**: "Dr." should stay together
- **Multi-word expressions**: "New York" as one token
- **Languages**: Different rules for different languages
- **Punctuation**: Handling quotes, hyphens, etc.

### Popular Tokenization Libraries
- **NLTK**: `nltk.word_tokenize()`, `nltk.sent_tokenize()`
- **spaCy**: Fast and accurate tokenization
- **Transformers**: Subword tokenizers (BERT, GPT)
- **Hugging Face Tokenizers**: High-performance tokenization

### Best Practices
- Choose tokenizer based on your NLP task
- Consider language-specific tokenizers
- Handle special cases (URLs, emails, numbers)
- Test tokenization on your specific domain text

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize, sent_tokenize
```

### Sentence tokenization

```python
sentences = "Her cat's name is Luna. Her dog's name is max"
sent_tokenize(sentences)
```

### Word tokenization

```python
sentence = "Her cat's name is Luna"
word_tokenize(sentence)
```

```python
sentence_2 = "Her cat's name is Luna and her dog's name is max"
word_tokenize(sentence_2)
```
