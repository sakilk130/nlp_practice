## 1.1 Lowercase

```python
sentence = "This is a complete sentence."
print(sentence)
```

```python
lower_sentence = sentence.lower()
print(lower_sentence)
```

```python
sentence_list = [
    "The sun rises in the east.",
    "I enjoy learning natural language processing.",
    "This notebook contains a simple text example.",
    "She reads a new book every month.",
    "We are practicing sentence preprocessing today.",
    "Python is a powerful programming language.",
    "He drinks coffee every morning.",
    "Machine learning can solve many problems.",
    "They went to the market after lunch.",
    "Text data often needs cleaning before analysis."
]
print(sentence_list)
```

```python
lower_sentence_list = [x.lower() for x in sentence_list]
print(lower_sentence_list)
```
