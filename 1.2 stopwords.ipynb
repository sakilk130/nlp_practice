## 1.2 Stopwords

### What are Stopwords?
Stopwords are common words that appear frequently in text but carry little meaningful information for text analysis tasks. Examples include: the, a, an, and, or, is, in, on, to, etc.

### Why Remove Stopwords?
- **Noise Reduction**: Removes frequent but less informative words
- **Improves Performance**: Reduces data dimensionality and processing time
- **Enhances Analysis**: Focuses on words with more semantic meaning
- **Better Tokenization**: Helps models learn more relevant patterns

### Common Stopwords Examples
- Articles: a, an, the
- Prepositions: in, on, at, to, from
- Conjunctions: and, or, but
- Pronouns: I, you, he, she, it
- Verbs: is, are, be, have

### When NOT to Remove Stopwords
- Sentiment analysis (e.g., "not good" vs "good")
- Negation-sensitive tasks
- Machine translation
- Named entity recognition

```python
# import packages

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
# stopwords
en_stopwords = stopwords.words('english')
print(en_stopwords)
```

```python
sentence = "Natural language processing is a fascinating field of artificial intelligence."
sentence_lower = sentence.lower()
```

```python
# remove stop words from sentence 
sentence_no_stopwords = ' '.join([word for word in sentence_lower.split() if word not in (en_stopwords)])
print(sentence_no_stopwords)
```

```python
# avoid to remove 
en_stopwords.remove("is")
en_stopwords.remove("not")
```

```python
# add custom words to stop words 
en_stopwords.append("go")
```

```python
sentance_no_stopwords_custom = ' '.join([word for word in sentence_lower.split() if word not in (en_stopwords)])
print(sentance_no_stopwords_custom)
```
