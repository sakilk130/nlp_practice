# 1.6 Lemmatization

## Why It Matters
Lemmatization reduces words to their base form (lemma) using linguistic knowledge. Unlike stemming, it produces valid dictionary words, making it more accurate for NLP tasks like sentiment analysis, classification, and information retrieval.

## Key Difference from Stemming
- **Stemming**: Removes suffixes algorithmically → "running" → "runn" (not a real word)
- **Lemmatization**: Uses vocabulary & morphology → "running" → "run" (valid word)

## Common Use Cases
- Text preprocessing for ML models
- Improving feature quality in NLP pipelines
- Better semantic understanding

```python
from nltk.stem import PorterStemmer
```

```python
connect_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connects']
learn_tokens = ['learned', 'learning', 'learn', 'learns', 'learner', 'learners']
likes_tokens = ['likes', 'better', 'worse']
```

```python
# create stemmer
ps = PorterStemmer()
```

```python
# connect tokens
for t in connect_tokens:
    print(t, " : ", ps.stem(t))
```

```python
# learn tokens
for t in learn_tokens:
    print(t, " : ", ps.stem(t))
```

```python
# like tokens 
for t in likes_tokens:
    print(t, " : ", ps.stem(t))
```

### Lemmatization

```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
# create lemmatizer 
lemmatizer = WordNetLemmatizer()
```

```python
# connect tokens 
for t in connect_tokens:
    print(t, " : ", lemmatizer.lemmatize(t))
```

```python
# learn tokens 
for t in learn_tokens:
    print(t, " : ", lemmatizer.lemmatize(t))
```

```python
# like tokens
for t in likes_tokens:
    print(t, " : ", lemmatizer.lemmatize(t))
```
